# Executive Summary
---

The goal of this project is to use various clustering methodologies to attempt to categorize different batter "types" from players during the 2023 Major Lague baseball season. This information will then be used to drive the development of a similarity matrix, which can provide valuable insight into the batters that compose a team. It can also be used to compare similar players from other teams, assisting in making informed player trades during trading windows. This can be a valuable asset to team managers. Player contracts can be hundreds of millions of dollars, and it is critical to understand a players true value in comparison to the rest of the player pool.

The Statcast data was first collected using the `pybaseball` library. Several aggregate statistics were computed that represent a wide range of batting metrics. The data frame was then filter to only include players that contributed a statistical meaning amount of data in the 2023 season.

While both accurately cluster players for n=2, the K-means revealed a third cluster that offered additional performance insights. 

Incorporating the cluster assignments generated from the previous step, a similarity matrix was computed using two methodologies. The first was a brute-force pairwise method that was computationally intensive and would not scale well if the data was expanded to more seasons. The second used a matrix vectorization method which significantly improved compute times. Examples on how to use the matrix are provdided to highlight the value proposition to team managers.

### Build a Summary Stats Dataframe

```python
# build lists of all teams and all batters (by player ID)
team_codes = stats['home_team'].unique().tolist()
batter_codes = sorted(stats['batter'].unique().tolist())
```

```python
# identifies the batters team based on the inning they are batting in
# home teams always bat second (bottom of the inning)
stats['batter_team'] = stats.apply(lambda row: row['home_team'] if row['inning_topbot'] == 'Bot' else row['away_team'], axis=1)

# create a list for summary stats dataframe * don't need this*
team = [stats[stats['batter']==batter]['batter_team'].iloc[0] for batter in batter_codes ]
```

```python
# calculate the total number of at-bats for each player in the season

# events that identify an at-bat 
# (AB does not include walks, sacrifice fly/sacrifice bunt, catcher interference, or hit by pitch)  
AB_events = ['strikeout', 'field_out', 'single', 'home_run', 'fielders_choice_out', 'fielders_choice',
             'double', 'force_out', 'grounded_into_double_play', 'triple', 'field_error', 'double_play',
             'strikeout_double_play', 'triple_play']

ABs = []
for batter in batter_codes:
    batter_AB = []
    tmp_df = stats[stats['batter'] == batter]

    for event in AB_events:
        batter_AB.append(len(tmp_df[tmp_df['events'] == event]))

    ABs.append(sum(batter_AB))  
```

```python
# The team each player ended the season on
team_eos = []
for batter in batter_codes:
    tmp_df = stats[stats['batter'] == batter].sort_values(by='game_date')
    
    team_eos.append(tmp_df.iloc[-1]['batter_team'])
```

```python
# determine the primary field position of the batter
# identify the position each player occupied the most

positions_list = ['pitcher.1', 'fielder_2.1', 'fielder_3',
                  'fielder_4', 'fielder_5', 'fielder_6',
                  'fielder_7', 'fielder_8', 'fielder_9',]
position = []
for batter in batter_codes:
    positions_played = []

    for i, pos in enumerate(positions_list):
        positions_played.append(stats[pos].eq(batter).sum())

    most_played_position = pd.Series(positions_played).idxmax()
    
    if most_played_position == 0: 
        position.append('pitcher')
    elif most_played_position == 1:
        position.append('catcher')
    elif most_played_position == 2:
        position.append('first_base')
    elif most_played_position == 3:
        position.append('second_base')
    elif most_played_position == 4:
        position.append('third_base')
    elif most_played_position == 5:
        position.append('short_stop')
    elif most_played_position == 6:
        position.append('left_field')
    elif most_played_position == 7:
        position.append('center_field')
    elif most_played_position == 8:
        position.append('right_field')
    
    
```

#### Computed Stats

The following stats are computed and are used to measure a players batting performance. These stats were chosen for their ability to capture a wide range of batting performance and for their strong predictive ability

* **wOBA** - Weighted on-base Average. This is a strong metric for determining how a batter contributes to offensive events. It is similar to the traditional Batting Average (BA). BA only takes into account whether a ball was hit, but is naive to the result of the hit. wOBA attmpts to weight events such as home-runs or triples higher than a single or field-out.
*  **BABIP** - Batting Average on Balls In Play. Measures how often batted balls are hit into the field. This is also similar to traditional Batting Average, but home-runs are not included, focusing on batters who excel at getting on base.
*  **ISO** - Isolated Power. Captures the frequency of batted balls that result in higher end hits such as triples and home runs. A player with high ISO would be considered a "power" hitter, and would likely bat 4th in a lineup to bring home as many runners as possible.
*  **Barrells per BBE** - Barrells per Batted Ball Event. Computes what percentage of batted balls are "barrels". A barrel is a ball that is struck on the "sweet spot" of the bat and launched at a high speed. This is determined by the launch angle (between 8 and 32 degrees) and exit velocity of over 95 MPH
*  **Average Launch Angle** - the average angle at which the ball leaves the players bat. Indicates how frequently the batter hits the ball in the "sweet spot" between 8 and 32 degrees. (batting accuracy)
*  **Average Launch Speed** - An indicator of how hard on average a player hits the ball. (batting power)

```python
# compute necessary batting stats for each player for the season
woba = []
babip = []
iso = []
barrels_per_bbe = []
avg_launch_angle = []
avg_launch_speed = []

babip_events = ['single', 'double', 'triple', 'field_out', 'fielders_choice_out',
                'fielders_choice', 'force_out', 'grounded_into_double_play', 'double_play',
                'field_error', 'sac_fly', 'sac_fly_double_play', 'strikeout', 'triple_play']
iso_events = ['single', 'double', 'triple', 'home_run', 'strikeout', 'field_out',
              'fielders_choice_out', 'fielders_choice', 'force_out', 'grounded_into_double_play',
              'double_play', 'strikeout_double_play', 'triple_play']
bbe_events = ['single', 'double', 'triple', 'home_run', 'field_out',
              'double_play', 'triple_play', 'fielders_choice', 'fielders_choice_out',
              'force_out', 'grounded_into_double_play', 'sac_bunt',
              'sac_bunt_double_play', 'sac_fly', 'sac_fly_double_play',
              'field_error', 'catcher_interf', 'fan_interference']

for batter in batter_codes:
    tmp_df = stats[stats['batter'] == batter]

    # WOBA
    woba.append(tmp_df['woba_value'].sum() / tmp_df['woba_denom'].sum())

    # BABIP
    babip_denom = tmp_df['events'].isin(babip_events).sum()
    babip.append(tmp_df['babip_value'].sum() / babip_denom)

    # ISO
    iso_denom = tmp_df['events'].isin(iso_events).sum()
    iso.append(tmp_df['iso_value'].sum() / iso_denom)

    # Barrels per batted-ball event
    bbes_df = tmp_df[tmp_df['events'].isin(bbe_events)]
    num_of_barrels = len(bbes_df[bbes_df['launch_speed_angle']==6])
    num_of_bbes = len(bbes_df)

    if num_of_bbes:
        barrels_per_bbe.append(num_of_barrels / num_of_bbes)
    else:
        barrels_per_bbe.append(0)

    # Average launch angle
    avg_launch_angle.append(bbes_df['launch_angle'].mean())

    # Average launch speed
    avg_launch_speed.append(bbes_df['launch_speed'].mean())

    
```

```python
# Create a dataframe with all of the computed information
batter_stats_df = pd.DataFrame({'batter': batter_codes, 'team': team_eos,
                               'position': position, 'at-bats': ABs, 
                               'wOBA': woba, 'BABIP': babip,
                               'ISO': iso, 'barrels_per_BBE': barrels_per_bbe,
                               'avg_launch_angle': avg_launch_angle,
                               'avg_launch_speed': avg_launch_speed})
```

## K-means

```python
def cluster_search_kmeans(df, nclusts=(2,10), n_init=10):
    ## If there are cluster assignments in the data frame remove them. 
    if 'cluster_assignments' in df.columns: df.drop(columns='cluster_assignments', inplace=True)
    WCSS=[]
    BCSS=[]
    silhouette=[]
    CH_index = []
    ## Compute total sum of squares
    x_bar = np.mean(df, axis=0)
    TSS = np.sum(np.sum(np.square(df - x_bar)))
    for n in range(nclusts[0],nclusts[1]+1):   
        temp_model = KMeans(n_clusters=n, n_init=n_init).fit(df)
        WCSS.append(temp_model.inertia_)
        BCSS.append(TSS - temp_model.inertia_)
        assignments = temp_model.predict(df)
        silhouette.append(silhouette_score(df, assignments))
        CH_index.append(calinski_harabasz_score(df, assignments))
    _, ax = plt.subplots(2,2, figsize=(8,8))    
    ax = ax.flatten()
    ax[0].plot(range(nclusts[0],nclusts[1]+1),WCSS)   
    ax[0].set_xlabel('Number of clusters')
    ax[0].set_ylabel('WCSS')
    ax[1].plot(range(nclusts[0],nclusts[1]+1),BCSS)   
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('BCSS')
    ax[2].plot(range(nclusts[0],nclusts[1]+1),silhouette)   
    ax[2].set_xlabel('Number of clusters')
    ax[2].set_ylabel('Silhouette Score')
    ax[3].plot(range(nclusts[0],nclusts[1]+1),CH_index)   
    ax[3].set_xlabel('Number of clusters')
    ax[3].set_ylabel('Calinski Harabasz Score')
 
nr.seed(9966)
cluster_search_kmeans(umap_df)    
```

The results of the clustering metrics strongly indicate the best grouping method for this data would be to use 2 clusters.

```python
def plot_cluster_assignments(df, cluster_assignments):
    df['cluster_assignments'] = cluster_assignments.astype('category')
    fig,ax = plt.subplots(figsize=(7,6))
    sns.scatterplot(data=df, x='component1', y='component2', 
                    hue='cluster_assignments', style='power_hitter', 
                    markers=['o','^',], s=25, alpha=0.6, ax=ax);
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    ax.set_xlabel('component 1');
    ax.set_ylabel('component 2');
    ax.set_title('UMAP projection of cluster assignments for batter dataset');

```

# 4. Similarity Matrix
---

With the knowledge gained from the cluster analysis, we can use the cluster assignments to further inform the similarity between batters. The goal of this step is to identify similar players. This information can be used in a variety of ways in combination with meta data such as salary to inform the trading of one player for another to improve team composition.

I first empolyed a brute-first pairwise comparison. This was very slow, memory intensive, and would not scale well if the comparisons were run across multiple seasons. The final version uses a matrix representaion to vectorize the similarity calculations.

```python
# batter_stats_df used from k-means clustering
batter_stats_df['cluster_assignment'] = batter_embedding_df['cluster_assignments']
batter_stats_df.head()
```

```python
def create_similarity_matrix_bf(df):
    euclidean = []
    cosine = []
    batter1_list = []
    batter2_list = []
    team1_list = []
    team2_list = []
    position1_list = []
    position2_list = []
    player1_cluster_list = []
    player2_cluster_list = []

    columns = ['wOBA', 'BABIP', 'ISO', 'barrels_per_BBE', 'avg_launch_angle', 'avg_launch_speed', 'cluster_assignment']
    
    batter_list = df.batter.unique().tolist()

    for pairs in combinations(batter_list, 2):
        batter1 = df[df['batter']==pairs[0]]
        batter2 = df[df['batter']==pairs[1]]

        batter1_stats = batter1[columns].values.reshape(1, -1)
        batter2_stats = batter2[columns].values.reshape(1, -1)

        # compute similarity and add it to each list
        euclidean.append(euclidean_distances(batter1_stats, batter2_stats)[0][0])
        cosine.append(cosine_similarity(batter1_stats, batter2_stats)[0][0])

        # add other features to their lists
        batter1_list.append(pairs[0])
        batter2_list.append(pairs[1])
        team1_list.append(batter1['team'].iloc[0])
        team2_list.append(batter2['team'].iloc[0])
        position1_list.append(batter1['position'].iloc[0])
        position2_list.append(batter2['position'].iloc[0])
        player1_cluster_list.append(batter1['cluster_assignment'].iloc[0])
        player2_cluster_list.append(batter2['cluster_assignment'].iloc[0])

        
    similarity_df = pd.DataFrame({'batter1': batter1_list, 'batter2': batter2_list,
                                  'team1': team1_list, 'team2': team2_list,
                                  'position1': position1_list, 'position2': position2_list,
                                  'player1_cluster': player1_cluster_list, 'player2_cluster': player2_cluster_list,
                                  'euclidean': euclidean, 'cosine': cosine})
    return similarity_df
```

```python
%time create_similarity_matrix_bf(batter_stats_df)
```

```python
def create_similarity_matrix(df):
    columns = ['wOBA', 'BABIP', 'ISO', 'barrels_per_BBE', 'avg_launch_angle', 'avg_launch_speed', 'cluster_assignment']
    
    # Extract feature matrix
    feature_matrix = df[columns].values

    # Compute pairwise distances
    euclidean_dist = euclidean_distances(feature_matrix)
    cosine_sim = cosine_similarity(feature_matrix)

    # Create a DataFrame to store results
    num_batters = len(df)
    batter1_list = []
    batter2_list = []
    team1_list = []
    team2_list = []
    position1_list = []
    position2_list = []
    player1_cluster_list = []
    player2_cluster_list = []
    euclidean = []
    cosine = []

    for i in range(num_batters):
        for j in range(i + 1, num_batters):
            batter1_list.append(df.iloc[i]['batter'])
            batter2_list.append(df.iloc[j]['batter'])
            team1_list.append(df.iloc[i]['team'])
            team2_list.append(df.iloc[j]['team'])
            position1_list.append(df.iloc[i]['position'])
            position2_list.append(df.iloc[j]['position'])
            player1_cluster_list.append(df.iloc[i]['cluster_assignment'])
            player2_cluster_list.append(df.iloc[j]['cluster_assignment'])
            euclidean.append(euclidean_dist[i, j])
            cosine.append(cosine_sim[i, j])

    similarity_df = pd.DataFrame({'batter1': batter1_list, 'batter2': batter2_list,
                                  'team1': team1_list, 'team2': team2_list,
                                  'position1': position1_list, 'position2': position2_list,
                                  'player1_cluster': player1_cluster_list, 'player2_cluster': player2_cluster_list,
                                  'euclidean': euclidean, 'cosine': cosine})
    })

    return similarity_df
```

```python
%time similarity_df = create_similarity_matrix(batter_stats_df)
similarity_df.head(10)
```

The second algorithm using the vectorized approach is significantly faster thatn the brute-force pairwise function

If the Yankees decide that they want to exhange one of there cluster 1 (power hitters) for a cluster 2 non power hitter, they can search by smilarity in a different cluster.
